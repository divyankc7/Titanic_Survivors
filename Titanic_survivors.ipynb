```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin'], axis = 1)
data
```

```python
x = data.drop(['Survived'], axis = 1)
y = data['Survived']
```

```python
x = pd.get_dummies(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_pred, y_test)
accuracy
```

```python
from sklearn.metrics import precision_score, f1_score, recall_score, confusion_matrix
```

```python
cm = confusion_matrix(y_pred, y_test)
print('Confusion_matrix: ')
print(cm)
```

```python
precision = precision_score(y_pred, y_test)
recall = recall_score(y_pred, y_test)
F1_score = f1_score(y_pred, y_test)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1_score: {F1_score:.2f}")
```

# Tuning the model

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
# n_estimators = number of DT to build
# max depth is the number of levels of DT
# min_samples_split is the minimum number of samples/obs to split an internal node
# min_samples_leaf is the minimum number of samples that are required to be at a leaf node
# bootstrap = True means that we're doing sampling with replacement; False means sampling without replacement
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = RF, param_grid = param_grid, cv = 5, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(xdt_train, ydt_train)
```

```python
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

print(f"Best parameters: {best_params}")
```

```python
y_best_pred = best_estimator.predict(xdt_test)
```

```python
best_accuracy = accuracy_score(ydt_test, y_best_pred)
print(f"Best accuracy: {best_accuracy:.2f}")
```

```python
Best_CM = confusion_matrix(y_best_pred, ydt_test)
Best_CM
```

```python
Recall = recall_score(y_best_pred, ydt_test)
Precision = precision_score(y_best_pred, ydt_test)
F1_score = f1_score(y_best_pred, ydt_test)

print(f"Recall: {Recall:.2f}")
print(f"Precision: {Precision:.2f}")
print(f"F1_score: {F1_score:.2f}")
```
